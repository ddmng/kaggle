```python
import pandas as pd
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
from sklearn.manifold import TSNE
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
train = pd.read_csv('../input/train.csv')
test = pd.read_csv('../input/test.csv')
```

```python
train.describe()
```

```python
X_train = train.iloc[:,2:]
y_train = train.iloc[:,1]
y_test = test.iloc[:,1:]
```

```python
clf = LogisticRegression(class_weight='balanced', penalty='l1', C=0.1, solver='liblinear').fit(X_train, y_train)
```

```python
predicts = clf.predict(X_train)
np.sqrt((predicts - y_train.values)*(predicts - y_train.values)).sum()/len(predicts)
```

```python
test_predicts = clf.predict(y_test)
test_predicts = pd.DataFrame(test_predicts, columns=['target'])
```

```python
test_predicts['id'] = test['id']
```

```python
test_predicts.to_csv('submission.csv', index=False)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(max_depth=8, min_samples_leaf=2, max_features=5)
clf = clf.fit(X_train,y_train)
predicts = clf.predict(X_train)
np.sqrt((predicts - y_train.values)*(predicts - y_train.values)).sum()/len(predicts)
```

```python
test_predicts = clf.predict(y_test)
test_predicts = pd.DataFrame(test_predicts, columns=['target'])
test_predicts['id'] = test['id']
test_predicts.to_csv('submission.csv', index=False)
```
