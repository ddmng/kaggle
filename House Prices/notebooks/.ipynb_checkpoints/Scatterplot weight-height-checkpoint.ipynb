```python
import pandas as pd
import matplotlib.pyplot as plt

wh = pd.read_csv("/home/daniele/Scaricati/weight-height.csv")
#wh = wh[wh['Gender']=='Male']
def color (row):
    if row['Gender'] == 'Male' :
        return 'c'
    else:
        return 'm'

wh['color']= wh.apply (lambda row: color (row),axis=1)

plt.figure(figsize=(15, 15))
ax = plt.subplot(1, 1, 1)
ax.scatter(wh['Weight']/2.20462, wh['Height']*0.0254*100, c=wh['color'],s=3,  alpha=.3)
plt.xlabel('Weight (kg)', fontsize=12)
plt.ylabel('Height (cm)', fontsize=12)
plt.grid()

ax.plot(91, 178, "ob")
ax.plot(59, 167, "or")
ax.plot(34, 137, "or")
ax.plot(33, 147, "ob")
plt.show()
```
